# 2. Lendo um arquivo de dados

Vamos carregar uma base utilizando o Pandas.


```python
arquivo = r"C:\Users\Joao.Maria\Documents\curso-python_joao_pedro\Base\Base_analitica_vendas.xlsx"

base_vendas = pd.read_excel(arquivo, sheet_name="Base")
```

Agora a variável **base_vendas** contém toda a tabela carregada.

# 4. Estrutura do Pandas

O Pandas possui duas estruturas principais:

| Estrutura | Descrição |
|---|---|
| **Series** | Uma única coluna |
| **DataFrame** | Uma tabela com várias colunas |


## 4.1 Series (uma coluna)

Uma **Series** representa apenas uma coluna da base.


```python
serie_clientes = base_vendas["DESCRICAOCLIENTE"]

serie_clientes
```

Verificando o tipo da estrutura:


```python
type(serie_clientes)
```

## 4.2 DataFrame (tabela)

Um **DataFrame** representa uma tabela completa.


```python
type(base_vendas)
```

Também podemos criar um DataFrame selecionando colunas específicas.


```python
dataframe_clientes = base_vendas[["DESCRICAOCLIENTE"]]

dataframe_clientes
```

Verificando o tipo:


```python
type(dataframe_clientes)
```
