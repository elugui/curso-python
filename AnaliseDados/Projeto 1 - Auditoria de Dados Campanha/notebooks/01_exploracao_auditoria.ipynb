# Projeto 1 - Auditoria de Dados de Campanha

Este notebook existe para o aluno **enxergar** o que o pipeline faz.

A lógica principal do projeto já está organizada em arquivos `.py`.
Aqui, o objetivo é:
1. carregar os dados
2. visualizar as bases
3. entender o tratamento da base bruta
4. executar as validações passo a passo
5. entender a saída final


```
import pandas as pd

from funcoes.funcoes_gerais import carregar_bases, preparar_base_bruta
from validacoes.regras_validacao import (
    validar_volume_apurado_x_bruto,
    validar_premiacao,
)
```

## 1. Carregar as bases

Aqui o aluno aprende que cada arquivo tem um papel:
- campanha: mostra o resultado apurado
- configuração: define quem e o que participa da campanha
- bruto: mostra o dado original transacional


```
bases = carregar_bases()

for nome, df in bases.items():
    print(f"{nome}: {df.shape}")
```

```
bases["relatorio_campanha_detalhe"].head()
```

## 2. Preparar a base bruta

Essa etapa filtra a base bruta para deixar somente o que interessa para a campanha.
O conceito importante aqui é: **nem todo dado bruto participa da regra de negócio**.


```
dados_bruto = preparar_base_bruta(
    relatorio_bruto=bases["relatorio_bruto"],
    configuracao_campanha_cliente=bases["configuracao_campanha_cliente"],
    configuracao_campanha_produto=bases["configuracao_campanha_produto"],
)

dados_bruto.head()
```

## 3. Validar volume apurado x volume bruto

Conceito:
- a campanha diz quanto foi apurado
- a base bruta mostra o dado original
- a validação compara os dois


```
resultado_volume = validar_volume_apurado_x_bruto(
    relatorio_campanha_detalhe=bases["relatorio_campanha_detalhe"],
    dados_bruto=dados_bruto,
)

resultado_volume.head()
```

## 4. Validar premiação

Conceito:
- a premiação depende do crescimento
- o crescimento gera um percentual de premiação
- esse percentual gera um valor
- depois comparamos o cálculo com o relatório oficial


```
resultado_premiacao = validar_premiacao(
    relatorio_campanha_premiacao=bases["relatorio_campanha_premiacao"],
    relatorio_campanha_detalhe=bases["relatorio_campanha_detalhe"],
)

resultado_premiacao.head()
```

## 5. O que o aluno deve entender

O projeto inteiro se resume a esta sequência:

**entrada → tratamento → regra → comparação → resultado**

Quando o aluno entende essa sequência, ele para de achar que pandas é só juntar `merge`, `groupby` e `apply` aleatoriamente.
